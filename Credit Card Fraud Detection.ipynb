```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
#distribution of legit & fraudulent transactions
credit_card_data['Class'].value_counts()
```

this dataset is highly unbalanced:
 0=>legit transaction
 1=>fraudulent transaction

```python
#seperating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transaction
credit_card_data.groupby('Class').mean()
```

Building a sample dataset containing similar distribution of legit and fraud transactions(called under-sampling).
No. of fraud transaction=492

```python
legit_sample=legit.sample(n=492)
```

```python
#concatenating 2 dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

we have made here 2 groups based on the 2 possible result(fraud/valid).

```python
#splitting the data into features and target
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#split the data into training & testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model training
model=LogisticRegression()
```

we have used logistic regression model here for evaluation because here the available dataset is of binary type in nature.

```python
model.fit(x_train,y_train)
```

```python
#Model evaluation
#1.Accuracy of training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
#1.Accuracy of testing data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Training data:',testing_data_accuracy)
```

Here training_data accurcy is almost similar to testing_data_accuracy & for both the data accuracy is greater than 90% which implies our model is very good.
